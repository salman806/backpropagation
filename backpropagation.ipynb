```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x1 = 0.1
x2 = 0.4
target = 0.7
learning_rate = 0.01
w1 = np.random.randn()
w2 = np.random.randn()
b = np.random.rand()
predict_output = []
network_error = []
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-1*x))

def error(predicted, target):
    return np.power(predicted - target,2)

def error_predicted_derive(predicted, target):
    return 2*(predicted - target)

def sigmoid_derive(x):
    return sigmoid(x)*(1-sigmoid(x))
    
def w_derive(x):
    return x

def update_weight(w, grade, learning_rate):
    return w - learning_rate*grade

    
    
```

```python
for k in range(80000):
    #forward pass
    y = x1*w1+x2*w2+b
    predicted = sigmoid(y)
    err = error(predicted, target)
    
    predict_output.append(predicted)
    network_error.append(err)
    
    #Backward pass
    g1 = error_predicted_derive(predicted,target)
    g2 = sigmoid_derive(y)
    
    g3w1 = w_derive(x1)
    g3w2 = w_derive(x2)
   
    
    gradw1 = g3w1*g2*g1
    gradw2 = g3w2*g2*g1
    
    w1 = update_weight(w1,gradw1,learning_rate)
    w2 = update_weight(w2,gradw2,learning_rate)
    
    #print(predicted)
    
```

```python
plt.plot(network_error)
plt.title('Iteration number VS prediction')
plt.xlabel('Iteration number')
plt.ylabel('Error')
plt.show()

plt.plot(predict_output)
plt.title('Iteration number VS prediction')
plt.xlabel('Iteration number')
plt.ylabel('prediction')
plt.show()

```
